```python
import sqlite3
from datetime import datetime, timedelta
from random import uniform

conn = sqlite3.connect('AdvancedSQLLab3.db')
cursor = conn.cursor()

commands = [
    """
    CREATE TABLE IF NOT EXISTS Temperature (
        Timestamp DATETIME NOT NULL,
        Temperature NUMERIC NOT NULL
    );
    """,
    """
    CREATE UNIQUE INDEX IF NOT EXISTS idx_timestamp ON Temperature (Timestamp);
    """
]

for command in commands:
    cursor.execute(command)

def dt(days):
    return timedelta(days=days)

N_rows = 500
now = datetime.now()

for i in range(N_rows):
    timestamp = now - dt(N_rows - i)
    temperature = uniform(18, 26)
    cursor.execute("INSERT INTO Temperature VALUES (?, ?)", (timestamp, temperature))

r_list = cursor.execute("SELECT datetime(Timestamp) as Date, Temperature FROM Temperature")

for rows in r_list:
    print(rows)

conn.commit()
conn.close()
import sqlite3
conn=sqlite3.connect('AdvancedSQLLab3.db')
cursor=conn.cursor()
r_list=cursor.execute("SELECT date('now','start of month', '+1 month', '-1 day')")
for rows in r_list:
  print(rows)
conn.commit()
conn.close()
```
